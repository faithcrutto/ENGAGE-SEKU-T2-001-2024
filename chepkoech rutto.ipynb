```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("BMI_train.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# Scatter plot
data.plot(kind='scatter', x="Weight", y="Height", s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#checking if there is missing data
data.isnull().sum()
```

```python
#Creating x and y values: X being the indepedent variable and y being the dependant variable

x=data[["Weight"]]
y=data[["Height"]]
```

```python
#creating the model
model=linear_model.LinearRegression()
```

```python
#splitting the data into 70% training and 30% testing
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
#training the model

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, label='Actual Values', color='grey', marker='o')

# Plot predicted values
plt.plot(x_test, y_pred, label='Predicted Values', color='gold', marker='x')
```

```python
model.intercept_
```

```python
model.coef_
```

```python
predicted_weigtht= model.predict([[10]])
print(predicted_weigtht)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("Mean squared error")
print(mean_squared_error(y_test,y_pred))
print("R_squared")
print(r2_score(y_test,y_pred))
print("Mean absolute error")
print(mean_absolute_error(y_test,y_pred))
```
